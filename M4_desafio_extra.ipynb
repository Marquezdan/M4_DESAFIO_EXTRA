```python
import pandas as pd

#Importação do data frame
dadosMilsa = pd.read_csv('http://www.leg.ufpr.br/~fernandomayer/data/milsa.csv')
dadosMilsa
```

```python
#Colocando a coluna Funcionario como indice da tabela:
dadosMilsa = dadosMilsa.set_index('Funcionario')
dadosMilsa
```

```python
#Retirada da coluna meses
dadosMilsa = dadosMilsa.drop('Meses',axis = 1)
```

```python
#Qual a média de idade dos funcionários? (Vou arredondar)
idade_media = dadosMilsa['Anos']
idade_media.mean().round(0)
```

```python
#Dos funcionários, com filhos, qual o número mais comum?
filhos = dadosMilsa[dadosMilsa['Filhos'] != 'NA']['Filhos'].value_counts().idxmax()
print(f'O número de filhos entre os funcionários mais comum é de {filhos} filhos.')
```

```python
#Qual a média e a mediana dos salários dos funcionários? O que esses valores podem indicar?
media_Salario = dadosMilsa['Salario'].mean()
mediana_Salario= dadosMilsa['Salario'].median()
print(media_Salario)
print(mediana_Salario)
```

```python
#Quais são os tipos de instrução existentes?
instrucoes = dadosMilsa['Inst'].unique()

instrucoesstg = 'Os tipos de instrução são ' +', '.join(instrucoes) +'.'
instrucoesstg
```

```python
#Qual a média de salário e idade das pessoas casadas?
casados_medias_Salário = dadosMilsa.loc[dadosMilsa['Est.civil'] == 'casado', 'Salario'].mean()
casados_medias_Idade = dadosMilsa.loc[dadosMilsa['Est.civil'] == 'casado', 'Anos'].mean()
print(f' A média de salários das pessoas casadas é {casados_medias_Idade} reais por hora.')
print(f' A média de idade dos casados é {casados_medias_Salário}.')
```

```python
 #Qual o funcionário que possui maior salário (informe seus atributos)?
indice_maior_salario = dadosMilsa['Salario'].idxmax()
func_maior_salario = dadosMilsa.loc[indice_maior_salario]
func_maior_salario
```

```python
#Qual o funcionário que possui menor idade (informe seus atributos)?
indice_mais_novo = dadosMilsa['Anos'].idxmin()
funcionario_mais_novo = dadosMilsa.loc[indice_mais_novo]
funcionario_mais_novo
```

```python
#Quem são os funcionários que possuem no máximo 35 anos, são da capital e tem instrução de 2o grau?
localizar_funcionarios  = dadosMilsa.loc[(dadosMilsa['Anos'] <= 35) & (dadosMilsa['Regiao'] == 'capital') & (dadosMilsa['Inst'] == '2o Grau')]
localizar_funcionarios
```

```python
#Faça uma correlação entre todos os atributos. (Dica: mapeie os atributos qualitativos para atributos quantitativos. Ex: solteiro: 1, casado: 2, etc).

#Convertendo os atributos qualitativos em quantitativos
estado_civil_remap = {'solteiro': 1, 'casado': 2}
inst_remap = {'1o Grau': 1, '2o Grau': 2, 'superior': 3}
regiao_remap = {'capital': 1, 'interior': 2, 'outro': 3}


dadosMilsa['Est.civil'] = dadosMilsa['Est.civil'].map(estado_civil_remap)
dadosMilsa['Inst'] = dadosMilsa['Inst'].map(inst_remap)
dadosMilsa['Regiao'] = dadosMilsa['Regiao'].map(regiao_remap)

correlacao = dadosMilsa.corr() 

correlacao
```

```python
#Pergunta extra 1: Existe correlação entre número de filhos e Anos?
print(dadosMilsa[['Filhos', 'Anos']].corr())
print('Como observado Existe uma correlação forte entre os atributos Filhos e Anos, indicando que eles possivelmente se influenciam')
```

```python
# Pergunta extra 2: Existe correlação entre o salário e Instrução?
print(dadosMilsa[['Inst', 'Salario']].corr())
print('Pode se observar uma correlação de quase 50%, indicando que sim, bem possivelmente esses atributos se influenciam')
```

```python
# Pergunta extra 3: Existe correlação entre a Instrução e a região?
print(dadosMilsa[['Inst', 'Regiao']].corr())
print('Com uma correlação negativa de pouco mais de um porcento, observa-se que existe uma correlação baixa entre o nível de instrução e a região')
```
